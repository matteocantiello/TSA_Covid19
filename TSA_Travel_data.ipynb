```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

import pandas as pd 
from matplotlib.patches import Rectangle
 
# Define some colors
BLACK      = (0.00,0.00,0.00)
ORANGE     = (0.90,0.60,0.00)
SKY_BLUE   = (0.35,0.70,0.90)
BLUE_GREEN = (0.00,0.60,0.50)
YELLOW     = (0.95,0.90,0.25)
BLUE       = (0.00,0.45,0.70)
VERMILLION = (0.80,0.40,0.00)
RED_PURPLE = (0.80,0.60,0.70)
```

```python
####################################################
# Plt Options from Perna, Lazzati & Cantiello 2017 #
####################################################
plt.rcParams['figure.figsize'] = (6,7)
plt.rcParams['font.family'] = ['Times New Roman']
#plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams['mathtext.fontset'] = 'dejavusans'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['mathtext.rm'] = 'serif'
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['axes.titlepad'] = 20 
plt.rcParams['legend.frameon'] = False
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['font.size'] = 16
plt.rcParams['legend.borderpad'] = 0.2
#plt.rc('legend', borderpad=0.2, markerscale=15., fancybox=False)
plt.rc('legend', borderpad=0.2, markerscale=1., fancybox=False)
plt.rcParams['legend.labelspacing'] = 0.5
plt.rcParams['legend.handletextpad'] = 0.5
plt.rcParams['legend.handlelength'] = 1
plt.rcParams["legend.framealpha"] = 1.0
####################################################
```

## Read the data

```python
file='Travel_data.csv'
import pandas as pd 
df = pd.read_csv(file, sep=',', parse_dates=['Date'],index_col=['Date'])           # Read csv file into a panda dataframe
```

## Plot Total Traveler Throughput (2019 and 2020)

```python
fig, ax = plt.subplots(figsize=(10, 10))
# Add x-axis and y-axis
ax.plot(df.index.values,
        df['Travel_2020']/1e6,
        color=BLUE_GREEN,lw=3)

ax.plot(df.index.values,
        df['Travel_2019']/1e6,
        color=SKY_BLUE,lw=3)

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Total Traveler Throughput (Millions)",
       title="TSA checkpoint travel numbers for 2020 and 2019")

# Rotate tick marks on x-axis
plt.setp(ax.get_xticklabels(), rotation=45)
```

## Plot relative fraction 

```python
fig, ax = plt.subplots(figsize=(10, 10))
# Add x-axis and y-axis

ax.plot(df.index.values,
        df['Travel_2020']/df['Travel_2019'],
        color=ORANGE,lw=5)

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Travelers 2020/2019",
       title="TSA checkpoint fractional travelers: 2020/2019")

# Rotate tick marks on x-axis
plt.setp(ax.get_xticklabels(), rotation=45)
```

## Make Two Panels Plot

```python
plt.rcParams['figure.figsize'] = (10,12)
xylim=[5.2,3.88]
lw=2
gs1 = gridspec.GridSpec(2, 1,height_ratios=[2,1])
# No space between plots, e.g.: gs1.update(left=0.1, right=0.9, wspace=0.00,hspace=0.00)
gs1.update(wspace=0.00,hspace=0.00)
# Assign axis and remove xticks from top plot
ax1 = plt.subplot(gs1[0,0])
ax2 = plt.subplot(gs1[1,0])
#ax3 = plt.subplot(gs1[2,0])
ax1.set_xticks([])

ax1.plot(df.index.values,
        df['Travel_2020']/1e6,
        color=BLUE_GREEN,lw=4,label='2020')

ax1.plot(df.index.values,
        df['Travel_2019']/1e6,
        color=SKY_BLUE,lw=4,label='2019')

ax1.set(ylabel="Total Traveler Throughput (Millions)",
       title="TSA checkpoint travel numbers for 2020 and 2019")

ax1.legend(loc='lower right',handlelength=2,fontsize=18)

ax2.plot(df.index.values,
        df['Travel_2020']/df['Travel_2019'],
        color=ORANGE,lw=4)
    
ax2.set(xlabel="Date",
       ylabel="Relative Fraction (2020/2019)")

ax2.annotate(r'Source: https://www.tsa.gov/coronavirus/passenger-throughput',xy=(0.32,0.33 ), xycoords='figure fraction', fontsize=16)
ax2.grid()
# Rotate tick marks on x-axis
#plt.setp(ax.get_xticklabels(), rotation=45)
plt.savefig("TSA_Covid19.pdf",bbox_inches='tight')
```
